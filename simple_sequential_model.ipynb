# First deep learning model using Keras

## Hypothetical situation - 

There was a obesity survey conducted on 2000 people,
The entire subject of survey are divided into two equal groups -
* Group A- age between 5-18
* Group B- age between 19-100

In group-A 5% of them are obese, 95% are not 

In group-B 13% of them are obese, while 87% of them are not.

```python
import numpy as np
from random import randint
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler
```

```python
train_labels=[]
train_samples=[]
```

## Generating the training sample data
We are using randint function to generate random ages and assinging their lables as 0, if not obese and 1, if obese.

### Group-A

```python
for i in range(100):
    random_adult = randint(19,100)
    train_samples.append(random_adult)
    train_labels.append(1)
    
for i in range(1900):
    random_adult = randint(5,18)
    train_samples.append(random_adult)
    train_labels.append(0)
```

### Group-B 

```python
   
for i in range(240):
    random_adult = randint(5,18)
    train_samples.append(random_adult)
    train_labels.append(1)

for i in range(1740):
    random_adult = randint(5,18)
    train_samples.append(random_adult)
    train_labels.append(0)
```

### Converting sample and target into numpy arrays

```python
train_labels= np.array(train_labels)
train_samples=np.array(train_samples)
train_labels,train_samples= shuffle(train_labels,train_samples)
for i in train_labels:
  print(i)
```

### Scaling down of the sample data

```python
scaler =MinMaxScaler(feature_range=(0,1))
scaled_train_samples= scaler.fit_transform(train_samples.reshape(-1,1))
for i in scaled_train_samples:
  print(i)
```

## Generating the testing sample data


```python
test_labels=[]
test_samples=[]
```

### Group-A

```python
for i in range(20):
    random_young = randint(19,100)
    test_samples.append(random_young)
    test_labels.append(1)
    
for i in range(380):
    random_adult = randint(5,18)
    test_samples.append(random_adult)
    test_labels.append(0)
```

### Group-B 

```python
   
for i in range(48):
    random_adult = randint(5,18)
    test_samples.append(random_adult)
    test_labels.append(1)

for i in range(348):
    random_adult = randint(5,18)
    test_samples.append(random_adult)
    test_labels.append(0)
```

### Converting sample and target into numpy arrays

```python
test_labels= np.array(test_labels)
test_samples=np.array(test_samples)
test_labels,test_samples= shuffle(test_labels,test_samples)
```

### Scaling down of the sample data

```python
scaler =MinMaxScaler(feature_range=(0,1))
scaled_test_samples= scaler.fit_transform(test_samples.reshape(-1,1))
for i in scaled_train_samples:
  print(i)
```

### Imports for Keras Sequential model

```python
import tensorflow as tf
from tensorflow import keras 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
```

```python
model = Sequential([
                Dense(units=16, input_shape=(1,), activation='relu'),
                Dense(units=32, activation='relu'),
                Dense(units=2, activation='softmax')    
])
```

```python
model.summary()
```

### prepare model for training

```python
model.compile(optimizer=Adam(learning_rate=0.0001),loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

### **Training**

```python
model.fit(x=scaled_train_samples,y=train_labels, validation_split=0.1, batch_size=10 ,epochs=30, shuffle=True, verbose=2)
```

### **Predict**

```python
predictions = model.predict(x=scaled_test_samples, batch_size=10, verbose=0)
```

```python
for i in predictions: 
  print(i)
```

```python
rounded_predictions= np.argmax(predictions, axis=-1)
for i in rounded_predictions:
  print(i)
```

### **Confusion Matrix**



```python
%matplotlib inline
from sklearn.metrics import confusion_matrix
import itertools
import matplotlib.pyplot as plt
```

```python
cm = confusion_matrix(y_true=test_labels, y_pred= rounded_predictions)
print(cm)
labels = ['not obese', 'obese']
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(cm)
plt.title('Confusion matrix of the classifier')
fig.colorbar(cax)
ax.set_xticklabels([''] + labels)
ax.set_yticklabels([''] + labels)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```
